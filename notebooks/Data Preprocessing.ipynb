```python

from pathlib import Path
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

DATASET_ROOT = Path(r"B:\Forth year\First semester\Machine\Assignments\dataset")

CLASSES = ["cardboard", "glass", "metal", "paper", "plastic", "trash"]

SAMPLES_PER_CLASS = 3
np.random.seed(42)

assert DATASET_ROOT.exists(), f"Dataset root not found: {DATASET_ROOT}"
print
```

```python

present = [c for c in CLASSES if (DATASET_ROOT / c).exists()]
missing = [c for c in CLASSES if not (DATASET_ROOT / c).exists()]

print("Present:", present)
print("Missing:", missing)

if missing:
       raise FileNotFoundError(f"Missing folders: {missing}")
```

```python

def list_images(folder: Path):
    exts = ".jpg"
    return [p for p in folder.glob("**/*") if p.suffix.lower() in exts and p.is_file()]

counts = {}
file_index = {}
for c in CLASSES:
    paths = list_images(DATASET_ROOT / c)
    counts[c] = len(paths)
    file_index[c] = paths

print("Image counts:")
for c in CLASSES:
    print(f"{c:<10}: {counts[c]}")
print("Total:", sum(counts.values()))

```

```python

paths = file_index["cardboard"]
sel = list(np.random.choice(paths, min(3, len(paths)), replace=False))

plt.figure(figsize=(12,4))
for i, p in enumerate(sel):
    img = Image.open(p).convert("RGB")
    plt.subplot(1, 3, i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title("cardboard")
plt.show()
```
